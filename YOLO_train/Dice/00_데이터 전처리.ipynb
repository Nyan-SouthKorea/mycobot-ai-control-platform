# 데이터 랜덤 스플릿

```python
import os
import random
import shutil

# 기본 경로 설정
path = './'
path_dataset = f'{path}/dataset'
path_origin = f'{path_dataset}/0_origin'

path_split = f'{path_dataset}/1_split'


# 이미지 리스트 불러오기
img_list = os.listdir(f'{path_origin}/images')

# 랜덤으로 섞기
random.shuffle(img_list)

# split 진행하기: 100장은 train, 나머지 20장은 val로 이동
# 폴더 생성
for mode in ['train', 'val']:
    for mode2 in ['images', 'labels']:
        os.makedirs(f'{path_split}/{mode}/{mode2}')

# 복사하기
for i, img_name in enumerate(img_list):
    # train/val 모드 설정
    if i < 100:
        mode = 'train'
    else:
        mode = 'val'
    
    # 레이블 이름 설정
    label_name = f"{img_name.split('.')[0]}.txt"

    # 이미지 복사
    from_path = f'{path_origin}/images/{img_name}'
    to_path = f'{path_split}/{mode}/images/{img_name}'
    shutil.copy(from_path, to_path)

    # 레이블 복사 and class 이름을 class_no 로 변경
    from_path = f'{path_origin}/labels/{label_name}'
    to_path = f'{path_split}/{mode}/labels/{label_name}'
    # 이름 변경
    with open(from_path, 'r', encoding='utf-8') as f:
        f = f.read()
    label_txt = f.replace('dice', '0')
    # 새로 저장
    with open(to_path, 'w', encoding='utf-8') as f:
        f.write(label_txt)
```

# 라벨 없는 이미지에 빈 텍스트 만들기
